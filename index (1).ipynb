Ahora nos fijamos que debemos de eliminar ese signo de dolar

```python
def clean_numeric(reserves, reserves_US):
    # Elimina el signo de dólar y las comas, y convierte a número
    reserves[reserves_US] = reserves[reserves_US].replace({'\$': '', ',': ''}, regex=True)
    reserves[reserves_US] = pd.to_numeric(reserves[reserves_US], errors='coerce')
    return reserves
```

```python
reserves = clean_numeric(reserves, "reserves_US")
```

Ahora luego de examinar, se procede a unir los data frame.

```python
for df in [reserves, energy, electricity, education]:
    df['country'] = df['country'].str.strip().str.lower()
    df['region'] = df['region'].str.strip().str.lower()
```

```python
reserves_sel = reserves[['country', 'region', 'reserves_US', 'reserves_yearData']]
energy_sel = energy[['country', 'region', 'Btu_energy', 'energy_yearData']]
electricity_sel = electricity[['country', 'region', 'kW_electricity', 'electricity_yearData']]
education_sel = education[['country', 'region', '%_PBI_education', 'education_yearData']]
```

```python
df_merged = reserves_sel.merge(energy_sel, on=['country', 'region'], how='outer')
df_merged = df_merged.merge(electricity_sel, on=['country', 'region'], how='outer')
df_merged = df_merged.merge(education_sel, on=['country', 'region'], how='outer')
```

```python
df_merged
```

Ahora solo ponemos los 176 paises que hay en común en las 4 tablas.

```python
df_merged = reserves_sel.merge(energy_sel, on=['country', 'region'], how='inner')
df_merged = df_merged.merge(electricity_sel, on=['country', 'region'], how='inner')
df_merged = df_merged.merge(education_sel, on=['country', 'region'], how='inner')
```

```python
df_merged
```

```python
# Agrupar por región y calcular promedio, min y max para cada variable
stats_by_region = df_merged.groupby('region').agg({
    'reserves_US': ['mean', 'min', 'max'],
    'Btu_energy': ['mean', 'min', 'max'],
    'kW_electricity': ['mean', 'min', 'max'],
    '%_PBI_education': ['mean', 'min', 'max']
})
# Mostrar resultado
stats_by_region
```

Ahora realizamos las etiquetas para las nuevas columnas

```python
# Definimos etiquetas
etiquetas = ['Muy bajo', 'Bajo', 'Medio', 'Alto', 'Muy alto']

# Creamos columnas categorizadas usando etiquetas
df_merged['reserves_level'] = pd.qcut(df_merged['reserves_US'], q=5, labels=etiquetas)
df_merged['energy_level'] = pd.qcut(df_merged['Btu_energy'], q=5, labels=etiquetas)
df_merged['electricity_level'] = pd.qcut(df_merged['kW_electricity'], q=5, labels=etiquetas)
df_merged['education_level'] = pd.qcut(df_merged['%_PBI_education'], q=5, labels=etiquetas)
cols_to_show=['country','region','reserves_US','Btu_energy','kW_electricity','%_PBI_education','reserves_level','energy_level','electricity_level','education_level']
```
